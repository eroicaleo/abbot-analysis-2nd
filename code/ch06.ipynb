```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 400)

# Define the function
y = x / (x**2 + 1)**2
z = x / (x**2 + 4)**2

# Plot the function
plt.plot(x, y, label='f(x) = x / (x^2 + 1)^2')
plt.plot(x, z, label='f(x) = x / (x^2 + 4)^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x / (x^2 + 1)^2')
plt.legend()
plt.grid(True)
plt.show()
```
